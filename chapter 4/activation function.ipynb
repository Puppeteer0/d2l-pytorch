```python
%matplotlib inline
import torch
from d2l import torch as d2l
```

ReLU(x) = max(x, 0).

```python
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)#保留正元素
y = torch.relu(x)
d2l.plot(x.detach(), y.detach(), 'x', 'relu(x)', figsize=(5, 2))
```

```python
y.backward(torch.ones_like(x), retain_graph=True)#backward假如不清零是累加的
d2l.plot(x.detach(), x.grad, 'x', 'grad of relu', figsize=(5, 2))
x.grad.zero_()
print("-------------------------------------------------------")
```

sigmoid(x) = 1/(1 + exp(−x))

```python
y = torch.sigmoid(x)
d2l.plot(x.detach(), y.detach(), 'x', 'sigmoid(x)', figsize=(5, 2))
```

```python
# 清除以前的梯度
x.grad.data.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
d2l.plot(x.detach(), x.grad, 'x', 'grad of sigmoid', figsize=(5, 2))
```

tanh(x) = (1 − exp(−2x))/(1 + exp(−2x)).

```python
y = torch.tanh(x)
d2l.plot(x.detach(), y.detach(), 'x', 'tanh(x)', figsize=(5, 2))
```

```python
# 清除以前的梯度
x.grad.data.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
d2l.plot(x.detach(), x.grad, 'x', 'grad of tanh', figsize=(5, 2))
```
